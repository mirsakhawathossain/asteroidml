```python
import numpy as np
import pandas as pd
import sklearn.datasets
from scipy.sparse import coo_matrix

import openml
from openml.datasets.functions import create_dataset
```

```python
data=pd.read_csv('train.csv')
```

```python
data.head(5)
```

```python
dataset=data.drop(columns=['id','pdes','epoch_mjd','epoch_cal','equinox','neo'])
```

```python
df = pd.DataFrame(dataset)
# enforce the categorical column to have a categorical dtype
df['pha'] = df['pha'].astype('category')
df['class']=df['class'].astype('category')
df['diameter']=df['diameter'].astype('float64')
print(df.info())
```

```python
pha_dataset=create_dataset(
    name="Asteroid_Dataset",
    description='Asteroid Dataset',
    creator='NASA, JPL',
    contributor=None,
    collection_date='12-02-2020',
    language='English',
    licence=None,
    default_target_attribute='pha',
    row_id_attribute='spkid',
    ignore_attribute=None,
    citation=None,
    attributes='auto',
    data=df,
    version_label='1',
    original_data_url=(
        'https://ssd.jpl.nasa.gov/'
    ),
)
```

```python
data_id =pha_dataset.publish()
print("Uploaded to https://www.openml.org/d/" + str(data_id))
```
